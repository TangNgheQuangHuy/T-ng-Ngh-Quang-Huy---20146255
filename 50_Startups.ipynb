```python
# Start up
import pandas as pd
url = '50_Startups.csv'
data = pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model #hoi quy tuyen tinh
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Profit']),
          y = data['R&D Spend']) # day. du lieu
print('Hệ số intercept là:', model.intercept_)
print('Hệ số coeficient là:',model.coef_)
model.score(pd.DataFrame(data['Profit']),data['R&D Spend'])
```

```python
y_pred=model.predict(X=pd.DataFrame(data['Profit']))
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'], data['Profit']**2]).T #Chuyển hàng thành cột
model.fit(X, y = data['R&D Spend'])
print('Intercept của mô hình:', model.intercept_)
print('Coefficient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X,y=data['R&D Spend'] ))
```

```python
import numpy as np
range = np.arange(0,200000,10)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
