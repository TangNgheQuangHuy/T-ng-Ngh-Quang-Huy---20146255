```python
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model #hoi quy tuyen tinh
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Salary']),
          y = data['YearsExperience']) # day. du lieu
print('Hệ số intercept là:', model.intercept_)
print('Hệ số coeficient là:',model.coef_)
```

```python
model.score(pd.DataFrame(data['Salary']),data['YearsExperience'])
```

```python
y_pred=model.predict(X=pd.DataFrame(data['Salary']))
```

```python
data.plot(kind='scatter',x='Salary',y='YearsExperience',figsize=(9,9),color='black')
plt.plot(data['Salary'],y_pred, color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2]).T 
model.fit(X, y = data['YearsExperience'])
print('Intercept của mô hình:', model.intercept_)
print('Coefficient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X,y=data['YearsExperience'] ))
```

```python
import numpy as np
range = np.arange(40000,130000,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='Salary',y='YearsExperience',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
