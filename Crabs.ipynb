```python
import pandas as pd
url = 'crabs.csv'
data = pd.read_csv(url)
data.head()
```

```python
data['speies']= data['speies'].replace('O',1)
data['speies']= data['speies'].replace('B',0)
```

```python
data['sex']= data['sex'].replace('M',1)
data['sex']= data['sex'].replace('F',0)
```

```python
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
x = data.loc[:,['speies','index','fontallip','rearwidth','length','width','depth']]
y =data['sex']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3)
sc=StandardScaler()
sc.fit(x_train)
x_train_Std=sc.transform(x_train)
x_test_Std=sc.transform(x_test)
```

```python
print(x_train,x_train_Std)
```

```python
model = Perceptron(max_iter=1000, eta0 = 0.1, random_state = 0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(y_pred)
print(y_test)
print('Do chinh xac ma hinh chua Scale:', accuracy_score(y_test, y_pred))
```

```python
model = Perceptron(max_iter=1000, eta0=0.1,random_state=0)
model.fit(x_train_Std,y_train)
y_pred=model.predict(x_test_Std)
print(y_pred)
print(y_test)
print('Do chinh xac mo hinh:',accuracy_score(y_test,y_pred))
```
